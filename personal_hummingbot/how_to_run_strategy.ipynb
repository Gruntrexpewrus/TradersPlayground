# How to run scripts on Hummngbot?

I am using Hyperliquid right now so there is need for connection.

Go here https://hummingbot.org/exchanges/hyperliquid/#i-exchange-info and pair your secret and public key. 
Verify connection

Now you want to select a strategy in your scripts and create a yaml file:

- Launch Hummingbot
    - docker compose up -d
    - docker attach hummingbot
    -  enter password
- run: create --script-config YOURSCRIPT and configure your script like sizes etc (select hyperliquid_perpetual_testnet as exchange)
- start --script YOURSCRIPT.py --conf YOURSCRIPT.yml
